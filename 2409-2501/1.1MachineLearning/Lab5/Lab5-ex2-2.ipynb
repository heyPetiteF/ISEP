```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#a) plot
beta= np.arange(-6,6,0.1)
R = np.sin(beta) + beta/10
plt.plot(beta,R, 'ro')
plt.ylabel('R')
plt.xlabel('beta')
```

```python
#b) derivative
# dR/dbeta = cos(beta)+(1/10)
```

```python
#c) gradiant descent with starting value for beta = 2.3
alpha = 0.1
beta_gd = [2.3]
for i in range(1,100):
    dbeta = np.cos(beta_gd[i-1])+(1/10)
    beta_gd.append(beta_gd[i-1]- alpha*dbeta)

plt.plot(beta_gd,"o")
print(beta_gd[99])
len(beta_gd) #4.612013
plt.ylabel('beta')
plt.xlabel('iteration')
```

```python
#d) gradiant descent with starting value for beta = 1.4
alpha = 0.1
beta_gd = [1.4]
for i in range(1,100):
    dbeta = np.cos(beta_gd[i-1])+(1/10)
    beta_gd.append(beta_gd[i-1]- alpha*dbeta)

plt.plot(beta_gd,"o")
print(beta_gd[99])
len(beta_gd) #-1.673957
plt.ylabel('beta')
plt.xlabel('iteration')
```
